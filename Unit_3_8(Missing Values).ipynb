```python
#missing values by replace()
s = pd.Series([0, 2, 3, 4, 5])
s.replace(0, 6)
```

```python
df = pd.DataFrame({'X': [0, 2, 3, 4, 5],
'Y': [6, 7, 0, 9, 1],
'Z': ['p', 'q', 'r', 's', 't']})
df.replace(0, 5)
df.replace([0, 2, 3, 4], 5)
```

```python
#to_replace
df = pd.DataFrame({'X': ['bbb', 'fff', 'bii'],
'Y': ['abc', 'bar', 'pqr']})

df.replace(to_replace=r'^bi.$', value='new', regex=True)
```

```python
import numpy as np
df = pd.DataFrame({'X': [np.NaN, 'fff', 'bii'],
'Y': ['abc', 'bar', 'pqr']})
df.replace(np.NaN,value=1)
```
